```python
# Import Meteostat library and dependencies
from datetime import datetime
import matplotlib.pyplot as plt
from meteostat import Point, Daily, Stations, Hourly, Monthly, Normals
```

```python
# Set time period
start = datetime(2021, 1, 1)
end = datetime(2021, 10, 17)

# Create Point for Adelaide, SA
#Meteostat: 94672
#National: /
#WMO: 94672
#ICAO: YPAD

adelaide = Point(-34.9333, 138.5167, 6)

# Get daily data for 2021
data = Daily(adelaide, start, end)
data = data.fetch()

# Plot line chart including average, minimum and maximum temperature
data.plot(y=['pres'])
plt.show()
```

```python
from meteostat import Stations

# Get nearby weather stations
stations = Stations()
stations = stations.nearby(-34.9333, 138.5167, 6)
station = stations.fetch(1)

# Print DataFrame
print(station)
```

```python
# Get hourly data
adelaide_airport = '94672'

data = Hourly(adelaide_airport, start, end)
data = data.fetch()

# Print DataFrame
print(data)
```

```python
data.plot(y=['pres'])
plt.show()
```

```python
data.tail(24)
```

```python
# Get Normals data
#data = Normals(adelaide_airport, 1990, 2020)
data = Normals(adelaide_airport)
data = data.fetch()

# Plot line chart including average, minimum and maximum temperature
data.plot(y=['pres'], figsize=(9,7))
plt.show()
```
