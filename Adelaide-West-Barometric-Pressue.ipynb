```python
badvalue = '      '

dfWT = dfWT.loc[dfWT['Station level pressure in hPa'] != badvalue]
dfWT.shape
```

```python
number = 1 # number of zeroes to add
dfWT['MMS'] = dfWT['MM'].astype('str')
dfWT['DDS'] = dfWT['DD'].astype('str')
dfWT['HHS'] = dfWT['HH24'].astype('str')
dfWT['MIS'] = dfWT['MI format in Local time'].astype('str')

dfWT['MMS'] = dfWT['MMS'].str.zfill(2)
dfWT['DDS'] = dfWT['DDS'].str.zfill(2)
dfWT['HHS'] = dfWT['HHS'].str.zfill(2)
dfWT['MIS'] = dfWT['MIS'].str.zfill(2)

dfWT['datesort'] = dfWT['Year Month Day Hour Minutes in YYYY'].astype('str') + dfWT['MMS'] + dfWT['DDS'] + dfWT['HHS'] + dfWT['MIS']
```

```python
dfWT['datesort']
```

```python
dfWT['datesortint'] = dfWT['datesort'].astype('int64')
```

```python
dfWT['datediff'] = dfWT['datesortint'].diff()
dfWT['pressurediff'] = dfWT['datesortint'].diff()
```

```python
dfWT['pressurediff'] = dfWT['Station level pressure in hPa'].diff()
```
